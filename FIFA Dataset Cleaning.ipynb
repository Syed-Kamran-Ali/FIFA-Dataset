## Importing Dataset for cleaning

```python
dataset = pd.read_csv("fifa21 raw data v2.csv",error_bad_lines=False, index_col=False, dtype='unicode')
dataset
```

```python
dataset.info()
```

### Stripping spaces from column names

```python
dataset.columns = dataset.columns.str.strip()
```

### Stripping Spaces from Club column 

```python
dataset["Club"] = dataset["Club"].str.strip()
```

### Making Replacing Function in order to replace the bad characters from different Columns

```python
def replace_bad_char(col_name,bad_char):
    for i in bad_char:
        dataset[col_name] = dataset[col_name].str.replace("i","")
```

```python
replace_bad_char("Name",["©","Ã","¼","Ä","™","‡","«","³","-","Å","±","-","‘","º","º","˜",""])
replace_bad_char("LongName",["©","Ã","¼","Ä","™","‡","«","³","-","Å","±","-","‘","º","º","˜",""])
```

```python
dataset.head()
```

### Sorting and Correcting datatype of Weight  Column

```python
dataset["Weight"].unique()
```

```python
for value,index in zip(dataset["Weight"],dataset["Weight"].index):
    if value.endswith("s"):
        a = value.strip("lbs")
        b = round(int(a)*0.453592)
        dataset["Weight"][index] = b
    elif value.endswith("kg"):
        c =value.strip("kg")
        dataset["Weight"][index] = int(c)
```

```python
dataset.rename({"Weight":"Weight(kg)"},axis = 1,inplace =True)
```

```python
dataset["Weight"] = dataset["Weight"].astype(int)
dataset["Weight"].dtype
```

### Sorting and Converting of Height Column

```python
dataset["Height"].unique()
```

```python
for value,index in zip(dataset["Height"],dataset["Height"].index):
    if value.endswith('"'):
        height_soughted = value.strip('"').split("'")
        height_cm = round((int(height_soughted[0])*30.48)+(int(height_soughted[1])*2.54))
        dataset["Height"][index] = height_cm
    elif value.endswith("cm"):
        c = value.strip("cm")
        dataset["Height"][index] = int(c)
```

```python
dataset.rename({"Height":"Height(cm)"},axis = 1,inplace =True)
```

```python
dataset["Height"] = dataset["Height"].astype(int)
dataset["Height"].dtype
```

## Correcting datatype of hits column

```python
dataset["Hits"] = dataset["Hits"].fillna("0")
```
